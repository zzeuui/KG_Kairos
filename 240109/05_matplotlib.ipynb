# 1. 맷플롯립의 구조

## 1.1 파이플롯

```python
x = range(100)
y = range(100)
plt.plot(x, y)
```

```python
import numpy as np

x_1 = range(100)
y_1 = [np.cos(value) for value in x]

x_2 = range(100)
y_2 = [np.sin(value) for value in x]

plt.plot(x_1, y_1)
plt.plot(x_2, y_2)
plt.show()
```

## 1.2 그림과 축

```python
fig, ax = plt.subplots()

x_1 = range(100)
y_1 = [np.cos(value) for value in x]

ax.plot(x_1, y_1)
ax.set(title='cos graph', xlabel='x', ylabel='y')

plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(10, 10)

ax_1 = fig.add_subplot(1, 2, 1)
ax_2 = fig.add_subplot(1, 2, 2)

ax_1.plot(x_1, y_1, c="b")
ax_2.plot(x_2, y_2, c="g")
plt.show()
```

## 1.3 서브플롯 행렬

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
print(ax)
print(type(ax))
```

```python
x=np.linspace(-1, 1, 100)
y_1 = np.sin(x)
y_2 = np.cos(x)
y_3 = np.tan(x)
y_4 = np.exp(x)

fig, ax = plt.subplots(2, 2)
ax[0, 0].plot(x, y_1)
ax[0, 1].plot(x, y_2)
ax[1, 0].plot(x, y_3)
ax[1, 1].plot(x, y_4)

plt.show()
```

```python
ax1 = plt.subplot(321) # (1) 첫 번째 공간에 axes 생성
plt.plot(x, y_1)
ax2 = plt.subplot(322) # (2) 두 번째 공간에 axes 생성
plt.plot(x, y_1)

####(312) = 3행 1열을 가지고 2번째에 위치함. = (323)+(324)
ax3 = plt.subplot(312) # (3) 두 번째 공간에 axes 생성
plt.plot(x, y_1)

ax4 = plt.subplot(325) # (4) 다섯 번째 공간에 axes 생성
plt.plot(x, y_1)
ax5 = plt.subplot(326) # (5) 여섯 번째 공간에 axes 생성
plt.plot(x, y_1)

plt.show()
```

# 2. 맷플롯립으로 그래프 꾸미기

## 2.1 색상

```python
X = range(100)
Y = range(100)

X_1 = range(100)
Y_1 = [value for value in X]
X_2 = range(100)
Y_2 = [value + 100 for value in X]
plt.plot(X_1, Y_1, color="#000000")
plt.plot(X_2, Y_2, c="c")
plt.show()
```

## 2.2 선의 형태

```python
plt.plot(X_1, Y_1, c="b", linestyle="dashed")
plt.plot(X_2, Y_2, c="r", ls="dotted")
plt.show()
```

## 2.3 제목

```python
plt.plot(X_1, Y_1, c="b", linestyle="dashed")
plt.plot(X_2, Y_2, c="r", ls="dotted")
plt.title("한글 폰트")
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(10,10)

ax_1 = fig.add_subplot(1,2,1)
ax_2 = fig.add_subplot(1,2,2)

ax_1.plot(X_1, Y_1, c="b")
ax_1.set_title("Figure 1")
ax_2.plot(X_2, Y_2, c="g")
ax_2.set_title("Figure 2")

plt.show()
```

## 2.4 범례

```python
plt.plot(X_1, Y_1,
color="b",
linestyle="dashed",
label='line_1')
plt.plot(X_2, Y_2,
color="r",
linestyle="dotted",
label='line_2')
plt.legend(
shadow=True,
fancybox=False,
loc="upper right")
plt.title('$y = ax+b$')
plt.xlabel('$x_line$')
plt.ylabel('y_line')
```

# 3. 맷플롯립에서 사용하는 그래프

## 3.1 산점도

```python
data_1 = np.random.rand(512, 2)
data_2 = np.random.rand(512, 2)
plt.scatter(data_1[:,0],
            data_1[:,1],
            c="b", marker="x")
plt.scatter(data_2[:,0],
            data_2[:,1],
            c="r", marker="o")
plt.show()
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)

colors = np.random.rand(N)
area = np.pi * ( 15 * np.random.rand(N))**2
plt.scatter(x, y, s=area, c=colors, alpha=0.5)

plt.show()
```

## 3.2 막대그래프

```python
# (1) 데이터 생성
data = [[5., 25., 50., 20.],[4., 23., 51., 17],[6., 22., 52., 19]]
# (2) X 좌표 시작점
X = np.arange(0,8,2)
# (3) 3개의 막대그래프 생성
plt.bar(X + 0.00, data[0], color = 'b', width = 0.50)
plt.bar(X + 0.50, data[1], color = 'g', width = 0.50)
plt.bar(X + 1.0, data[2], color = 'r', width = 0.50)
# (4) X축에 표시될 이름과 위치 설정
plt.xticks(X+0.50, ("A","B","C", "D"))
# (5) 막대그래프 출력
plt.show()
```

## 3.3 누적 막대그래프

```python
data = np.array([[5., 25., 50., 20.], [4., 23., 51., 17], [6., 22., 52., 19]])
color_list = ['b', 'g', 'r']
data_label = ["A","B","C"]
X = np.arange(data.shape[1])
data = np.array([[5., 5., 5., 5.], [4., 23., 51., 17], [6., 22., 52., 19]])

for i in range(3):
  plt.bar(X, data[i], bottom = np.sum(data[:i], axis=0),
          color = color_list[i], label=data_label[i])

plt.legend()
plt.show()
```
